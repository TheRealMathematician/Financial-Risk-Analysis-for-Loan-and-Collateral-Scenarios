```python
import pandas_datareader as pdr
import datetime

# Set the start and end date for the data retrieval
start_date = datetime.datetime(2019, 9, 1)  # Start date
end_date = datetime.datetime(2023, 9, 1)    # End date

# Ticker symbols for 1-year Treasury constant maturity rate
# Replace with other FRED indicators if needed
tickers = ['GS1']

# Fetch the data for the specified tickers using pandas_datareader
data = pdr.get_data_fred(tickers, start_date, end_date)

print(data.head())
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
data.plot(ax=ax)
plt.ylabel('Percentage Rate')
plt.title('1-year US Treasury')
```
